```python
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
import numpy as np

# File to Load 
Cameron = "Resources/Cameron.csv"
Sullivan = "Resources/Sullivan.csv"

Cameron_df = pd.read_csv(Cameron)
Sullivan_df = pd.read_csv(Sullivan)
Cameron_df.head()
```

```python
Sullivan_df.head()
```

```python
# Combine the data into a single dataset
combined_data_df = Cameron_df.append(Sullivan_df)
combined_data_df
```

Get the count of voters by political parties declared.

Party Code abbreviations:
D = Democrat
I = Independent
LN = LIBERTARIAN
NF = NO AFFILIATION
NO = NO PARTY
NON = NONE
NOP = NON-PARTISAN
OTH = OTHER
R = REPUBLICAN

```python
# Use Group By to see the count of voters by Party Code
grouped_party_df = combined_data_df.groupby(['Party Code']).count()
party_df = grouped_party_df['ID Number']
party_df
```

```python
# Total number of declared parties as registered by the voters
total_parties = party_df.count()

# Total number of voters
sum_parties = party_df.sum()

print(f'Out of the {sum_parties} total voters in Sullivan and Cameron County, there were {total_parties} different political parties declared.')

# Total of non-affiliated, non-declared parties converted to a list
no_affiliation = party_df['NF'] + party_df['NO'] + party_df['NON'] + party_df['NOP'] 
no_affiliation_df = [no_affiliation]

# Total of all other parties converted to a list
others = sum_parties - no_affiliation - party_df['R'] - party_df['D'] - no_affiliation
others_df = [others]

# Number of Democrats converted to a list
democrat = party_df['D']
democrat_df = [democrat]

# Number of Republicans converted to a list
republican = party_df['R']

print(' ')

# Combine the lists into one DataFrame
party_pie_df = pd.DataFrame()
party_pie_df['No Affiliation'] = no_affiliation_df
party_pie_df['Others'] = others_df
party_pie_df['Democrats'] = democrat
party_pie_df['Republicans'] = republican

party_pie_df
```

```python
# Labels for the sections of our pie chart

grouped_party_df = combined_data_df.groupby(['Party Code']).count()
party_df = grouped_party_df['ID Number']
party_df

total_parties = party_df.count()
sum_parties = party_df.sum()

plt.title("% of Total Voters by Parties")
labels = ["No Affiliation", "Others", "Democrats", "Republicans"]
sizes = [533, 243, 2425, 4036]
colors = ["lightgreen", "skyblue", "yellow", "lightcoral"]
explode = (0.1, 0.1, 0.1, 0.1)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=120)
plt.tight_layout()
plt.axis("tight")
plt.savefig("% of Total Voters by Parties.png")
plt.show()
```

The next set of cells is an anlysis of voters by age group bins

```python
DOB = combined_data_df['DOB']
total = len(DOB)

# Python3 code to  calculate age in years. Returns the years in a list
from datetime import date
from datetime import datetime
years = []

for i in DOB:
    def calculateAge(i):
        time_in_datetime = datetime.strptime(i, "%m/%d/%Y")
    
        today = date.today()
        age= time_in_datetime.year
        #age = today.year - i.year - ((today.month, today.day) 
           # (i.month, i.day))
        return age
    k=calculateAge(i)
    years.append(k)

# Create a DataFrame of the DOB by Years only   
voters =[]
voters = pd.DataFrame(voters)
voters['Years'] = years
voters
```

```python
# binning by years
bins = [1900, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
bin_labels = ["<1919", "1920 - 1929", "1930 - 1939", "1940 - 1949", "1950 - 1959", "1960 - 1969", "1970 - 1979", "1980 - 1989", "1990 - 1999", "2000 - 2009"]

voters["Group by Years"] = pd.cut(voters["Years"], bins, labels=bin_labels)
voters.head()

age =voters.groupby('Group by Years').count()
print(age)

total_voters = age['Years'].sum()

total_voters
```

```python
age =voters.groupby('Group by Years').count()
total_voters = age['Years'].sum()

total_voters = [15, 171, 630, 1255, 1684, 1242, 850, 842, 598]
age = np.arange(len(total_voters))

plt.bar(age, total_voters, color='skyblue', alpha=0.95, align="center")

tick_locations = [value for value in age]
plt.xticks(age, ["<1919", "1920 - 1929-- (2%)", "1930 - 1939-- (8%)", "1940 - 1949-- (17%)", "1950 - 1959-- (23%)", "1960 - 1969-- (17%)", "1970 - 1979-- (11%)", "1980 - 1989-- (11%)", "1990 - 1999-- (8%)"])

plt.xlim(-0.95, len(age)-0.95)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, max(total_voters)+1700)
plt.axis("tight")
plt.title("Voters Age Demographics")
plt.xlabel("Voters by Date of Birth (Year)")
plt.ylabel("Number of Voters")
plt.savefig("Voters Age Demographics.png")
```
